# Deep Learning
## ANN - Artificial Neuronal Network

```python
import tensorflow as tf
import numpy as np
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
from keras.datasets import mnist
from keras import metrics
from sklearn.model_selection import train_test_split
print(tf.__version__)
```

# Regression

```python
from keras.datasets import boston_housing
(x_train_validation, y_train_validation), (x_test, y_test) = boston_housing.load_data()

x_train, x_validation, y_train, y_validation = train_test_split(x_train_validation, y_train_validation, test_size=0.2, random_state=42)

model = Sequential()
model.add(Dense(64, input_dim=13, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(1))

model.compile(loss='mean_squared_error', optimizer='adam')
history = model.fit(x_train, y_train, epochs=100, validation_data=(x_validation, y_validation), batch_size=32)
```

```python
test_acc = model.evaluate(x_test, y_test)
print('Test mse:', test_acc)
```

```python
plt.figure(figsize=(10, 8))
plt.plot(history.history['loss'], label='loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.title('Loss Function')
plt.legend()
plt.show()
```

```python
model.compile(loss='mean_absolute_error', optimizer='adam')
history = model.fit(x_train, y_train, epochs=100, validation_data=(x_validation, y_validation), batch_size=32)
```

```python
test_loss = model.evaluate(x_test, y_test)
print('Test mae:', test_loss)
```

```python
plt.figure(figsize=(10, 8))
plt.plot(history.history['loss'], label='loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.title('Loss Function')
plt.legend()
plt.show()
```

```python
import pickle

filename = 'regression_nn.pkl'
pickle.dump(model, open(filename, 'wb'))

from google.colab import files
files.download(filename)
```

# Image Classification

```python
from keras.datasets import fashion_mnist
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
import numpy as np
from keras.models import Sequential



(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()

print("Training data shape:", train_images.shape)
print("Training labels shape:", train_labels.shape)
print("First training label:", train_labels[0])

plt.imshow(train_images[0], cmap="gray")
plt.title(f"Label: {train_labels[0]}")
plt.axis("off")
plt.show()

train_images = train_images / 255.0
test_images = test_images / 255.0

train_images = train_images.reshape((60000, 28 * 28))
test_images = test_images.reshape((10000, 28 * 28))
```

```python
model = keras.Sequential()
model.add(Dense(128, activation='relu', input_shape=(784,)))
model.add(Dense(10, activation='softmax'))

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

history = model.fit(train_images, train_labels, epochs=5, batch_size=32, validation_split=0.33)
```

```python
test_loss, test_acc = model.evaluate(test_images, test_labels)
print("Test accuracy:", test_acc)
```

```python
plt.figure(figsize=(10, 8))
plt.plot(history.history['loss'], label='loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.title('Loss Function')
plt.legend()
plt.show()
```

```python
filename_img = 'image_classification_nn.pkl'
pickle.dump(model, open(filename_img, 'wb'))
files.download(filename_img)
```

# Text Classification

```python
from keras.datasets import imdb
from keras.models import Sequential
from keras.layers import Embedding, GlobalAveragePooling1D, Dense, Flatten, Dropout
from keras.preprocessing.sequence import pad_sequences

num_words = 50000
(x_train_validation, y_train_validation), (x_test, y_test) = imdb.load_data(num_words=num_words)

x_train, x_validation, y_train, y_validation = train_test_split(x_train_validation, y_train_validation, test_size=0.2, random_state=10)

x_train = pad_sequences(x_train, maxlen=256)
x_validation = pad_sequences(x_validation, maxlen=256)
x_test = pad_sequences(x_test, maxlen=256)

batch_size = 32
train_dataset = tf.data.Dataset.from_tensor_slices((x_train, y_train)).batch(batch_size)
validation_dataset = tf.data.Dataset.from_tensor_slices((x_validation, y_validation)).batch(batch_size)
test_dataset = tf.data.Dataset.from_tensor_slices((x_test, y_test)).batch(batch_size)

model = Sequential()
model.add(Embedding(input_dim = num_words, output_dim = 64, input_length=256))
model.add(GlobalAveragePooling1D())
model.add(Dense(64))
model.add(Dropout(0.6))
model.add(Dense(1,activation='sigmoid'))
model.compile(optimizer='rmsprop',
              loss='binary_crossentropy',
              metrics=['accuracy'])

with tf.device('/GPU:0'):
  history = model.fit(train_dataset, epochs=10, validation_data=validation_dataset)
```

```python
test_loss, test_acc = model.evaluate(test_dataset)
print('Test accuracy:', test_acc)
print('Test loss:', test_loss)
```

```python
plt.figure(figsize=(10, 8))
plt.plot(history.history['loss'], label='loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.title('Loss Function')
plt.legend()
plt.show()
```

```python
import pickle

filename_text = 'text_classification_nn.pkl'
pickle.dump(model, open(filename_text, 'wb'))

from google.colab import files
files.download(filename_text)
```
